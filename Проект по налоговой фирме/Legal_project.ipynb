**3.2 Рекомендации по проведению мероприятий**

Преимущественным спросом у юридических лиц являются **налоговые споры**.

Главная ориентация в продвижении предприятия должна быть направлена на

потребности клиента. Необходимо учитывать, что юридические услуги

**нематериальны**, что является основной проблемой в восприятии

клиентом. Нематериальность порождает страх. Страх потребителя выражается в


мыслях: «А что будет, если не понравится?», «А надежный ли юрист?»


и так далее.


Для решения задач нами разработаны рекомендации к разработке

мероприятий:

**Оффлайн мероприятия**

1.	Для привлечения внимания потенциальных и поддержания прочных связей

с текущими клиентами рекомендуется разработка и распространение

сувенирной продукции в качестве подарка для клиентов, в виде фирменных

календарей, блокнотов, визиток и ручек. **Корпоративные подарки** – это

инструмент способный улучшить взаимоотношения с клиентами. Подарки

используются для формирования корпоративного имиджа и деловой репутации,

укрепления связей с клиентами и деловыми партнерами.

2.	Для стимуляции и мотивации старых клиентов, а также повышения уровня

лояльности новых клиентов рекомендуется к разработке **система скидок**,

которая будет транслироваться на информационных ресурсах компании.

3.	**Участие в ежегодной конференции предпринимателей**, организованная

некоммерческой организацией палатой предпринимателей, где представляется

возможность показать себя с компетентной стороны, а мероприятие будет

освещаться в прессе.  Также это возможность для

распространения визитных карточек и сувенирной продукции компании.

**3.3 Расчёт ROI для коммуникационных мероприятий**

Путём подсчёта стоимости запланированных мероприятий компании для

проведения мероприятий необходимо затратить **170 000 у.е.**. на достижение

Целесообразно рассчитать эффективность кампании с помощью коэффициента

**ROI (Return on Investment)**. Этот показатель рассматривает рекламу как

инвестиции и подсчитывает рентабельность.


Согласно настройкам таргетированной и контекстной рекламы, наша целевая

аудитория, на которую мы ориентировались, составляет в среднем 4 239

организации.

1.	Подсчитаем стоимость одного контакта (затраты на кампанию / целевая

аудитория):


**170 000 / 4 239 = 40,1**

2.	Рассчитаем доход от коммуникационных мероприятий: так как продвижение

еще только проектируется, мы сделаем предположение о том, какое

Советник» и узнать, является ли это число достаточным для того, чтобы

считать кампанию успешной.

Допустим, по итогам коммуникационных мероприятий услугой юридического

консалтинга воспользовались организации 42 (1% ЦА) средняя стоимость

услуг юридического консалтинга в ООО «Статский Советник» составляет 20 000

рублей в месяц, таким образом:

**42 х 20 000 = 840 000 у.е..**

Эта сумма превышает среднюю выручку компании на одну четверть.

Рассчитаем коэффициент ROI (доход от кампании – затраты
